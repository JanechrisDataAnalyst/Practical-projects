INTRODUCTION:

Obesity is a disorder involving excessive body fat that increases the risk of health problems. Obesity often results from taking in more calories than are burned by exercise and normal daily activities. Obesity usually occurs when a person's body mass index is 30 or greater. The main symptom is excessive body fat, which increases the risk of serious health problems.
