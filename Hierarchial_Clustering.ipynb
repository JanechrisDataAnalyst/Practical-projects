```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading Dataset
df = pd.read_csv("/content/Mall_Customers.csv")
```

```python
df.head()
```

# DATA ANALYSIS PHASE

># Applying the ML Clustering Algorithm: Hierarchy as hc

```python
import scipy.cluster.hierarchy as hc
```

```python
# Predict Attributes:
x = df.iloc[:,[3,4]].values
```

```python
dendrogram = hc.dendrogram(hc.linkage(x, method  = "ward"))
plt.title(" Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidian Distance")
```

```python
# Build Hierarchial with Optimal No. Cluster
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 5, affinity = "euclidean", linkage = "ward")
model
```

```python
# fit the model with data
y_hc = model.fit_predict(x)
```

```python
plt.scatter(x[y_hc == 0,0], x[y_hc == 0,1], s=100,c = "blue", label = "Careful")
plt.scatter(x[y_hc == 1,0], x[y_hc == 1,1], s=100,c = "green",label = "Standard")
plt.scatter(x[y_hc == 2,0], x[y_hc == 2,1], s=100,c = "red", label = "Target")
plt.scatter(x[y_hc == 3,0], x[y_hc == 3,1], s=100,c = "cyan", label = "Careless")
plt.scatter(x[y_hc == 4,0], x[y_hc == 4,1], s=100,c = "magenta", label = "Sensible")
plt.title ("Cluster of Client")
plt.xlabel("Annual Income(k$)")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```
